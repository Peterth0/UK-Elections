```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_data_2024 = r"D:\thomson\Desktop\Statistics\Data\2024_Detailed_Constituency_ Results.xlsx"

data_2024 = pd.read_excel(path_data_2024)
data_2024 = data_2024.iloc[1:]  
data_2024.columns = data_2024.iloc[0]

data_2024 = data_2024[1:].reset_index(drop=True)
```

```python
data_2024.head()
```

```python
columns = data_2024.columns
print(columns)

numeric_columns = ['Electorate', 'Valid votes', 'Invalid votes', 'Majority', 'Con', 'Lab', 'LD', 'RUK', 'Green', 'SNP', 'PC', 'DUP',
       'SF', 'SDLP', 'UUP', 'APNI', 'All other candidates',
       'Of which other winner']

data_2024[numeric_columns] = data_2024[numeric_columns].apply(pd.to_numeric, errors='coerce')
#data_2024
```

```python
# Winning party - 'First party'
# Second place - 'Second party'

#'ONS region ID'
#'ONS ID'

region_data_2024 = data_2024.groupby('ONS region ID')[numeric_columns].sum().reset_index()
region_data_2024
```

```python
total_electorate = region_data_2024['Electorate'].sum()
total_electorate
```
